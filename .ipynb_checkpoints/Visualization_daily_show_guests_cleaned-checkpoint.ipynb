### Import Library

```python
import matplotlib.pyplot as plt
import pandas as pd
```

### Read .csv file

```python
df = pd.read_csv('daily_show_guests_cleaned.csv', encoding='cp1252')
```

### Filter by Year (2014)

```python
year_2014 = df[df['YEAR'] == 2014]
```

### Grouping and Counting Media

```python
grouping = year_2014.groupby('Group').count().reset_index()
```

### Pie Chart Visualization

```python
labels = grouping.Group
plt.pie(grouping['Total'], labels = labels, autopct='%.2f %%', radius = 2, textprops = {"fontsize": 12}, shadow = True)
plt.show()
```
