### Import Library

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Read .csv file

```python
df = pd.read_csv('bike_share.csv', encoding='cp1252')
```

### Filter by Year (2012)

```python
year_2012_initial = df[df.dteday >= '2012-01-01']
year_2012 = year_2012_initial[year_2012_initial.dteday <= '2012-12-31']
```

### Grouping by Month

```python
group_by_month_registered = year_2012.groupby('mnth').sum().registered.reset_index()
group_by_month_total_rentals = year_2012.groupby('mnth').sum().total_rentals.reset_index()
```

### Bar Chart Visualization

```python
xpos_registered = np.arange(len(group_by_month_registered.mnth))
xpos_total_rentals = np.arange(len(group_by_month_total_rentals.total_rentals))
plt.figure(figsize=(15,7))
plt.bar(xpos_registered-0.2, group_by_month.registered, width=0.4)
plt.bar(xpos_total_rentals+0.2, group_by_month_total_rentals.total_rentals, width=0.4)
plt.title('Registered and Total Rentals Graphic in 2012', fontdict={'fontname': 'Calibri', 'fontsize': 20})
plt.xlabel('Month', fontdict={'fontname': 'Calibri', 'fontsize': 16})
plt.ylabel('Total', fontdict={'fontname': 'Calibri', 'fontsize': 16})
plt.legend(['Total Registered', 'Total Rentals'])
plt.show()
```
