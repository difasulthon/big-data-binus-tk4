### Import Library

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.dates import date2num
```

### Read .csv file

```python
df = pd.read_csv('insurance_premiums.csv', encoding='cp1252')
```

### Filter Region (West)

```python
west = df[df['Region'] == 'West']
```

### Bar Chart Visualization

```python
xpos = np.arange(len(west.State))
plt.figure(figsize=(15,7))
plt.bar(xpos, west.fatal_collisions_speeding, width=0.2)
plt.bar(xpos+0.2, west.fatal_collisions_alc, width=0.2)
plt.bar(xpos+0.4, west.fatal_collisions_not_distracted, width=0.2)
plt.bar(xpos+0.6, west.fatal_collisions_no_hist, width=0.2)
plt.xticks(xpos, west.State)
plt.title('Fatal Collision Comparartion in West Region', fontdict={'fontname': 'Calibri', 'fontsize': 20})
plt.xlabel('State', fontdict={'fontname': 'Calibri', 'fontsize': 16})
plt.ylabel('Total', fontdict={'fontname': 'Calibri', 'fontsize': 16})
plt.legend(['Fatal Collisions Speeding', 'Fatal Collisions Alc', 'Fatal Collisions Not Distracted', 'Fatal Collisions No Hist'])
plt.show()
```
