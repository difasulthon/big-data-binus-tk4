### Import Library

```python
import matplotlib.pyplot as plt
import pandas as pd
```

### Read .csv file

```python
df = pd.read_csv('us-counties.csv', encoding='cp1252')
```

### Filter California State

```python
california = df[df['state'] == 'California']
```

### Filter Los Angeles County

```python
los_angeles = california[california['county'] == 'Los Angeles']
```

### Filter March Cases

```python
march_cases_start = los_angeles[los_angeles.date >= '2020-03-01']
march_cases = march_cases_start[march_cases_start.date <= '2020-03-31']
```

### Line Chart Visualization

```python
plt.figure(figsize=(15,5))
plt.plot(march_cases.date, march_cases.cases, linewidth = 2)
plt.plot(march_cases.date, march_cases.deaths, linewidth = 2)
plt.xticks(rotation = 'vertical')
plt.legend(['Jumlah Kasus', 'Kematian'])
plt.title('Pergerakan Jumlah Kasus dan Tingkat Kematian Pada Maret 2020 di Los Angeles', 
          fontdict={'fontname': 'Calibri', 'fontsize': 20})
plt.xlabel('Tanggal', fontdict={'fontname': 'Calibri', 'fontsize': 16})
plt.ylabel('Jumlah', fontdict={'fontname': 'Calibri', 'fontsize': 16})
plt.show()
```
